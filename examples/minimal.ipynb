```python
%matplotlib inline

from time import sleep
import numpy as np

from livelossplot import PlotLosses
```

```python
plotlosses = PlotLosses()

for i in range(10):
    plotlosses.update({
        'acc': 1 - np.random.rand() / (i + 2.),
        'val_acc': 1 - np.random.rand() / (i + 0.5),
        'loss': 1. / (i + 2.),
        'val_loss': 1. / (i + 0.5)
    })
    plotlosses.send()
    sleep(0.5)
```

```python
# start from epoch 2
# make it smaller

plotlosses = PlotLosses(from_step=2, figsize=(6, 3))

for i in range(10):
    plotlosses.update({
        'acc': 1 - np.random.rand() / (i + 2.),
        'val_acc': 1 - np.random.rand() / (i + 0.5),
        'loss': 1. / (i + 2.),
        'val_loss': 1. / (i + 0.5)
    })
    plotlosses.send()
    sleep(0.5)
```

```python
# only last 5 steps

plotlosses = PlotLosses(from_step=-5)

for i in range(10):
    plotlosses.update({
        'acc': 1 - np.random.rand() / (i + 2.),
        'val_acc': 1 - np.random.rand() / (i + 0.5),
        'loss': 1. / (i + 2.),
        'val_loss': 1. / (i + 0.5)
    })
    plotlosses.send()
    sleep(0.5)
```
