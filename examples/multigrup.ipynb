# livelossplot example: groups starting not simultaneously

```python
%matplotlib inline

from time import sleep
import numpy as np

from livelossplot import PlotLosses, MainLogger
from livelossplot.outputs import BaseOutput
from livelossplot.outputs import MatplotlibPlot, ExtremaPrinter
```

```python
groups ={}
for j in range(4):
    groups['acccuracy' + str(j)] = ['acc' + str(j), 'val_acc'+ str(j)]
for j in range(4):
    groups['log-loss' + str(j)] = ['loss'+ str(j), 'val_loss' + str(j)]
print(groups)
plotlosses = PlotLosses(groups=groups)
for epoch in range(100):
    id = epoch //25
    plotlosses.update({
        'acc' + str(id): 1 - np.random.rand() / (epoch + 2.),
        'val_acc' + str(id): 1 - np.random.rand() / (epoch + 0.5),
        'loss' + str(id): 1. / (epoch + 2.),
        'val_loss' + str(id): 1. / (epoch + 0.5)
    })
    plotlosses.send()
    sleep(.01)
```
